```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('data.csv')
df.head()
```

**Find mean for X**

```python
df.X.mean()
```

**Find mean for Y**

```python
df.Y.mean()
```

```python
model = LinearRegression()
```

```python
model.fit(df.X.array.reshape(-1,1), df.Y)
```

**b1 coefficient**

```python
round(model.coef_[0],2)
```

**b0 intercept**

```python
round(model.intercept_, 2)
```

**R^2**

```python
round(model.score(df.X.array.reshape(-1,1), df.Y), 2)
```

# Задача №2

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('candy-data.csv')
```

**Exclude from train data 'Hersheys Special Dark' and 'Lemonhead'**

```python
df_train = df[(df.competitorname != 'Hersheys Special Dark') & (df.competitorname != 'Lemonhead')]
x_train = df_train[df_train.columns[1:12]]
y_train = df_train.winpercent
```

```python
model = LinearRegression().fit(x_train, y_train)
```

**Predict winpercent for 'Hersheys Special Dark' and 'Lemonhead'**

```python
df_test = df[(df.competitorname == 'Hersheys Special Dark') | (df.competitorname == 'Lemonhead')]
x_test = df_test[df_test.columns[1:12]]
```

```python
[round(i, 3) for i in model.predict(x_test)]
```

**Predict winpercent for special candy**

```python
test_candy = [0, 1, 1, 0, 1, 1, 1, 0, 0, 0.012, 0.192]
```

```python
[round(i, 3) for i in model.predict([test_candy])]
```
