# Задание №1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
X=pd.read_csv('94_16.csv', header=None)
```

**Переход к новым координатам**

```python
pca = PCA(n_components=2, svd_solver='full')

# 1 var
X_centered = X - X.mean(axis=0)
pca.fit(X_centered)
X_pca = pca.transform(X_centered)
pd.DataFrame(X_pca).head()

# 2 var
X_pca2 = pca.fit_transform(X)
pd.DataFrame(X_pca2).head()
```

**Координаты первого объекта относительно главных компонент**

```python
[round(i, 3) for i in X_pca[0]]
```

**Доля объясненной дисперсии при использовании двух и трех главных компонент**

```python
exp = [round(pca.explained_variance_ratio_[i], 3) for i, component in enumerate(pca.components_)]
print(round(exp[0] + exp[1], 3))
print(round(sum(exp), 3))
```

**Выделение групп объектов при испоьзовании двух главных компонент**

```python
plt.scatter([n[0] for n in X_pca], [n[1] for n in X_pca])
```
